```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

>Step 1. Load the dataset


```python
sales=pd.read_csv('sales-data.csv')
sales.sample(5)
```

```python
sales.shape
```

```python
sales.isnull()
```

```python
sales.isna().sum()
```

>Step 2 - Prepare a NaN tabular data

```python
na_count=sales.drop(['Date','Temperature','Fuel_Price'],axis=1).isna().sum()
print(na_count)
```

>Step 3 - Plot the bar chart from the NaN tabular data, and also print values on each bar


```python
sb.barplot(na_count.index.values,na_count)
plt.xticks(rotation=90);

for i in range(na_count.shape[0]):
    count=na_count[i]
    
    plt.text(i,count+250,count,ha='center',va='top')
```
