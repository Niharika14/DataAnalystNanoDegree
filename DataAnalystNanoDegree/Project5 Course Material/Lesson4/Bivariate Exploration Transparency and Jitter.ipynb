```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

fuel_econ=pd.read_csv('fuel-econ.csv')

fuel_econ.head(3)
```

>Example 1. Jitter - Randomly add/subtract a small value to each data point


```python
plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
sb.regplot(data=fuel_econ, x='year', y='comb', 
           truncate=False)

plt.subplot(1,2,2)
sb.regplot(data=fuel_econ, x='year', y='comb', 
           truncate=False, x_jitter=0.3)
```

>Example 2. Plot with both Jitter and Transparency


```python
plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
sb.regplot(data=fuel_econ, x='year', y='comb',
          truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20});

plt.subplot(1,2,2)
sb.regplot(data=fuel_econ, x='year', y='comb',
          truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20});
plt.scatter(data = fuel_econ, x = 'year', y = 'comb', alpha=1/20);
```

```python
plt.figure(figsize=[20,5])

plt.subplot(1,4,1)
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', truncate=False, x_jitter=0.2, scatter_kws={'alpha':1/10});

plt.subplot(1,4,2)
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', truncate=False, x_jitter=0.2, scatter_kws={'alpha':1/20});

plt.subplot(1,4,3)
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/10});

plt.subplot(1,4,4)
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20});
```
